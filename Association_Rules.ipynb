```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
encoder= TransactionEncoder().fit(transactions)
```

```python
one_hot=encoder.transform(transactions)
```

```python
one_hot
```

```python
one_hot.shape
```

```python
df1=pd.DataFrame(one_hot,columns=encoder.columns_)
```

```python
df1
```

```python
df1.shape
```

# Apriori

```python
frequent_itemsets=apriori(df1,min_support=0.05,use_colnames=True)
```

```python
frequent_itemsets
```

# Association rules

```python
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0.05)
```

```python
rules
```

# Analysis and Interpretation:

```python
# support
#(spaghetti) >(mineral water) with high support means that a significant portion of customers are purchasing both together.
```

```python
# Lift gives an idea of the strength of the relationship between products compared to random chance.
# (spaghetti)>(mineral water) with lift 1.43 .5 means that chocolate and wine are bought together 1.43 times more often than would be expected if the purchases were independent.
```

```python
# confidence
# (spaghetti)>(mineral water) with high confidence of 0.34 means that 30% of the transactions that have spaghetti also have mineral water.
```

```python
# 1. What is Lift and Why is it Important in Association Rules?
```

```python
 # Lift measures the strength of the association between two items, compared to how often the items would occur together by chance.
#A lift greater than 1 means that A and B are more likely to appear together than by chance (positive association).
# Lift helps identify strong associations between items that are not obvious and can guide marketing strategies or store placement.
```

```python
# support:he proportion of transactions that contain a particular item or itemset.
# support(A)=(NumberÂ ofÂ transactionsÂ containingÂ A)/(Total number of transactions)

```

```python
# confidence:The likelihood that item B is purchased when item A is purchased.
# Confidence(Aâ†’B)= Support(Aâˆ©B)/Support(A)

```
